```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Load CSV file into dataframe and rename columns for clarity
df0 = pd.read_csv('heart.csv')
df = df0.rename(index=str, columns={"cp":"chest_pain_type", "trestbps":"rest_bp", "thalach":"max_hr", "exang":"exercise_ang"})

# Convert column names to a list and print
column_names = df.columns.tolist()
print("Column names:", column_names)

# Determine information about the dataframe
print("Columns:")
print(df.columns)
print("Shape:")
print(df.shape)
print("Head:")
print(df.head())

# Check for missing values
print("\nMissing values:")
print(df.isnull().sum())

# Check for duplicates, count, and output information
duplicates = df[df.duplicated()]
num_duplicates = duplicates.shape[0]
print("Duplicate rows: ", duplicates)
print(f"Total number of duplicate rows: {num_duplicates}")

# Identify only the first instance of each duplicate row
duplicates_first = df[df.duplicated(keep='first')]
print('\nOnly the first instance of each duplicate row:')
print(duplicates_first)

# Filter out unique rows and display
unique_rows = df[~df.duplicated()]
print("\nUnique rows: ", unique_rows)

# Drop duplicates
df_cleaned = df.drop_duplicates()
print("\nNumber of duplicate rows after cleaning:", df_cleaned.duplicated().sum())

#save cleaned data as a new CSV file
df_cleaned.to_csv(r'C:\Users\enapi\Captsone 2\Lib\heart_disease_clean.csv', index=False)



```
